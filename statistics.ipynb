```python


df = pd.DataFrame({'close': sample_data})
df['returns'] = df['close'] / df['close'].shift(1) - 1
df = df.dropna()

Q1 = df['returns'].quantile(0.25)
Q3 = df['returns'].quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5

df_no_outliers = df[(df['returns'] >= Q1 - threshold * IQR) & (df['returns'] <= Q3 + threshold * IQR)]

print(f"no outliers length {df_no_outliers['returns'].size}")

sample_data = df_no_outliers['returns'].to_numpy()

stat, p = stats.shapiro(sample_data)


hist, bins, _ = plt.hist(df['returns'], bins=20, edgecolor='black')  # 'Value' is the name of the column you added

last_edge = None

for count, bin_edge in zip(hist, bins):
    if count:
        print(f"count = {count}, edge_start = {last_edge} bin edge = {bin_edge}")
        last_edge = bin_edge


plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.hist(df_no_outliers['returns'], bins=20, edgecolor='black')  # 'Value' is the name of the column you added
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()


print(f"Shapiro-Wilk Statistic: {stat}")
print(f"P-value: {p}, less than 0.05 {p < 0.05}")
```
